```python
#ans-
import pandas as pd
data = {
    'User': ['A', 'B', 'C'],
    'Book1': [5, 4, 0],
    'Book2': [3, 0, 4],
    'Book3': [0, 4, 5]
}
df = pd.DataFrame(data)
from sklearn.metrics.pairwise import cosine_similarity

df.set_index('User', inplace=True)
sim = cosine_similarity(df)
similar_users = sim[0].argsort()[-2:][::-1]

recs = []
for idx in similar_users:
    for book in df.columns:
        if df.iloc[idx][book] > 0 and df.iloc[0][book] == 0:
            recs.append(book)

print("Recommended books for A:", set(recs))
```

```python
#2.
movies = pd.DataFrame({
    'Title': ['A', 'B', 'C'],
    'Description': ['Hero fights villains', 'Romantic comedy film', 'Action packed thriller']
})

#Question:
#Recommend movies similar to Movie A based on description.

#Expected Output:
#Most similar movie to A: C
```

```python
import pandas as pd
movies = pd.DataFrame({
    'Title': ['A', 'B', 'C'],
    'Description': ['Hero fights villains', 'Romantic comedy film', 'Action packed thriller']
})

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(movies['Description'])
cos_sim = cosine_similarity(tfidf_matrix)

similar_idx = cos_sim[0].argsort()[-2]
print("Most similar movie to A:", movies.iloc[similar_idx]['Title'])
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Print the first 5 rows of transformed data
print("First 5 rows after PCA transformation:")
print(X_pca[:5])

```

```python
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

data = load_wine()
X = StandardScaler().fit_transform(data.data)

pca = PCA(n_components=3)
pca.fit(X)
print(pca.explained_variance_ratio_)
```

```python
#5.
#Question:
#Select the top 2 features from Iris dataset using Mutual Information.

#Expected Output:
#Top 2 features: [2 3]
```

```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, mutual_info_classif
import pandas as pd

data = load_iris()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

selector = SelectKBest(score_func=mutual_info_classif, k=2)
selector.fit(X, y)
print("Top 2 features:", selector.get_support(indices=True))
```
