```python
#5.

from sklearn.feature_selection import SelectKBest, f_classif
import pandas as pd

df = pd.DataFrame({
    'age': [20, 30, 40],
    'gender': ['M', 'F', 'M'],
    'target': [1, 0, 1]
})

X = df[['age', 'gender']]
y = df['target']

selector = SelectKBest(score_func=f_classif, k=1)
X_new = selector.fit_transform(X, y)

print("Shape:", X_new.shape)
```

```python
#6.
import pandas as pd
movies = pd.DataFrame({
    'title': ['Inception', 'Interstellar', 'Dark Knight', 'Tenet'],
    'description': [
        'dream within a dream and sci-fi thriller',
        'space time travel and love across galaxies',
        'dark superhero action and Gotham city',
        'reversed time logic and spy action'
    ]
})
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
tfidf = TfidfVectorizer()
matrix = tfidf.fit_transform(movies['title'])  # Mistakenly used 'title'
similarity = cosine_similarity(matrix)
print(similarity)
```

```python
#9.
import pandas as pd

df = pd.DataFrame({
    'user': ['A', 'B', 'A', 'C'],
    'movie': ['X', 'Y', 'Y', 'X'],
    'rating': [4, 5, 3, 2]
})
top_movie = df.groupby('movie')['ratings'].mean().idxmax()
print("Top Movie:", top_movie)
```
